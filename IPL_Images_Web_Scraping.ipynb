```python
Team_Name=["chennai-super-kings","delhi-capitals","gujarat-titans","kolkata-knight-riders","lucknow-super-giants","mumbai-indians","punjab-kings","rajasthan-royals","royal-challengers-bangalore","sunrisers-hyderabad"]
```

```python
Player_Name=[]
Player_Img = []
for j in Team_Name:
    url =f"https://www.iplt20.com/teams/{j}/squad/2024#list"
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    for i in soup.find_all("div",class_="ih-p-img"):
        Player_Name.append(i.text.split("\n")[3])
        Player_Img.append(i.img)
```

```python
Player_Img
```

```python
Player_Name
```

```python
imgs=[]
for i in Player_Img:
    print(type(i))
    ab=str(i)
    da = ab.replace('<img alt="" class="lazyload" data-src="',"").replace('"/>',"")
    imgs.append(da)
```

```python
imgs
```

```python
da={"Player_Name":Player_Name,"imgs":imgs}
```

```python
import pandas as pd
```

```python
data=pd.DataFrame(da)
```

```python
data.to_csv("imgs.csv")
```
